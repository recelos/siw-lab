## Zadanie 1

Utwórz zdjęcie sytntetyczne jak na przykładzie i zapisz je na dysku.  
  
  

```python
import numpy as np

h = 500
w = 500

image = np.zeros((h, w))
image.fill(255)
start = 0
bar_width = 3
increment = 3

while start < w:
    for i in range(3):
        end_x = start + bar_width
        if end_x > w:
            end_x = w
    
        image[:, start:end_x] = 0
        start = end_x + bar_width

    bar_width += increment
    start+=increment
    increment*=2

plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

plt.imsave('Lab2zdjecia/out.png', image, cmap='gray')
```

## Zadanie 2

Bazując na utworzonym zdjęciu w zadaniu 1 napisz kod pozwalający na zamianę losowych pikseli na kolor biały (jeżeli kolor piksela jest czarny) i czarny (jeśli kolor piksela jest szary lub biały)

```python
from random import randint

def swap_pixels(image: cv.typing.MatLike, num: int) -> cv.typing.MatLike:
    swapped = []
    h, w, _ = image.shape
    for _ in range(num):
        x = randint(0, w-1)
        y = randint(0, h-1)
        while (x, y) in swapped:
            x = randint(0, w-1)
            y = randint(0, h-1)
            
        if all(image[x, y] == [0, 0, 0]): 
            image[x, y] = [255, 255, 255]
        else:
            image[x, y] = [0, 0, 0]

        swapped.append((x, y))
    return image

image = cv.imread('Lab2zdjecia/out.png')
new_img = swap_pixels(image, 5000)

plt.imshow(new_img, cmap='gray')
```
